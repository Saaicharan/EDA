```python
import pandas as pd
```

```python
data = pd.read_csv('/content/WA_Fn-UseC_-Telco-Customer-Churn.csv')
data.head()
```

```python
# prompt: How many men and women (sex feature) are represented in this dataset?

data['gender'].value_counts()
```

```python
# prompt: code for EDA with the above dataset

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of Churn
sns.countplot(x='Churn', data=data)
plt.title('Distribution of Churn')
plt.show()

# Churn by Gender
sns.countplot(x='gender', hue='Churn', data=data)
plt.title('Churn by Gender')
plt.show()

# Churn by Internet Service
sns.countplot(x='InternetService', hue='Churn', data=data)
plt.title('Churn by Internet Service')
plt.xticks(rotation=45)
plt.show()

# Distribution of Tenure
sns.histplot(data['tenure'], kde=True)
plt.title('Distribution of Tenure')
plt.show()

# Correlation Matrix
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np  # Importing NumPy

# Distribution of Churn
sns.countplot(x='Churn', data=data)
plt.title('Distribution of Churn')
plt.show()

# Churn by Gender
sns.countplot(x='gender', hue='Churn', data=data)
plt.title('Churn by Gender')
plt.show()

# Churn by Internet Service
sns.countplot(x='InternetService', hue='Churn', data=data)
plt.title('Churn by Internet Service')
plt.xticks(rotation=45)
plt.show()

# Distribution of Tenure
sns.histplot(data['tenure'], kde=True)
plt.title('Distribution of Tenure')
plt.show()

# Correlation Matrix
plt.figure(figsize=(12, 8))
# Excluding non-numeric columns from correlation calculation
sns.heatmap(data.select_dtypes(include=np.number).corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
